<h2>Queue</h2>
<p>Queue adalah sekumpulan data yang mana penambahan elemen hanya bisa dilakukan pada suatu ujung disebut dengan sisibelakang(rear), dan penghapusan(pengambilan elemen) dilakukan lewat ujung lain (disebut dengan sisi depan atau front).</p>

<p>Pada <b>Stack</b> atau tumpukan menggunakan prinsip <b>“Masuk terakhir keluar pertama” atau LIFO (Last In First Out)</b>, Maka pada <b>Queue</b> atau antrian prinsip yang digunakan adalah <b>“Masuk Pertama Keluar Pertama” atau FIFO (First In First Out)</b>.<br>
Queue atau antrian banyak kita jumpai dalam kehidupan sehari-hari, ex: antrian Mobil diloket Tol, Antrian mahasiswa Mendaftar, dll.<br>
Contoh lain dalam bidang komputer adalah pemakaian sistem komputer berbagi waktu(time-sharing computer system) dimana ada sejumlah pemakai yang akan menggunakan sistem tersebut secara serempak.<br>
Pada Queue atau antrian Terdapat satu buah pintu masuk di suatu ujung dan satu buah pintu keluar di ujung satunya dimana membutuhkan variabel Head dan Tail ( depan/front, belakang/rear).</p>

```python
class qeue:
    def __init__(self):
        self.items=[]
        
    def isempty(self):
        return self.items==[]
    
    def en(self,item):
        self.items.insert(0,item)
        
    def de(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
q=qeue()
q.en(4)
q.en('dog')
q.en(True)
print(q.size())
print(q.de())
```

<h2>DEQUEUE</h2>

<p>Dequeue hampir sama dengan Queue, Perbedaannya pada jalan masuk dan keluar.</p>
<p>Dequeue memiliki sisi depan dan sisi belakang, jadi data bisa masuk lewat depan atau belakang, dan keluar lewat depan atau belakang.</p>



```python
class qeue:
    def __init__(self):
        self.items=[]
        
    def isempty(self):
        return self.items==[]
    
    def addrear(self,item):
        self.items.insert(0,item)
        
    def removerear(self):
        return self.items.pop(0,item)

    def addfront(self,item):
        return self.items.append(item)

    def removefront(self):
        return self.items.pop()

    def size(self):
        return len(self.items)
    
q=qeue()
q.addrear('dono')
q.addrear('wawan')
q.addfront('tani')
q.size()
```
