<h2>SINGLE LINKED LIST</h2>

<p>Single Linked List merupakan sebuah tempat yang disediakan pada satu area memori tertentu untuk menyimpan data yang dikenal dengan sebutan node atau simpul. Setiap node memiliki pointer yang menunjuk ke simpul berikutnya sehingga terbentuk satu untaian, dengan demikian hanya diperlukan sebuah variabel pointer. Susunan berupa untaian semacam ini disebut Single Linked List. Biasanya Linked List pada node terakhir akan menunjuk ke NONE, dimana  NONE memilik nilai khusus yang artinya tidak menunjuk ke mana-mana.</p>



<h2>SEARCH</h2>
<p>Menggunakan variabel boolean Found, karna diawal kita masih belum menemukan item yang kita cari. Found bisa diset ke FALSE</p>

<p>Iterasinya:</p>
<p> Jika datanya = CURRENT NODE maka set variabel booleannya ke TRUE</p>
<p>Jika tidak. maka tetap lanjutkan ke node berikutnya</p>

Susunan berupa untaian model ini disebut Doubly Linked List. </p>
