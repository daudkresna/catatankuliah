<h2>INFiX</h2>
<p>Infix adalah cara penulisan atau ungkapan yang meletakan operator di tengah antara 2 operand dalam hal ini dalam kurung sangat menentukan posisi.</p>
<p>Contoh Infix :</p>
<p>A + B</p>
<p>( A - B ) * C</p>

<h2>PREFiX</h2>
<p>Prefix adalah cara penulisan atau ungkapan yang meletakan operator disebelah kiri 2 operand dan dalam kurung sangat menentukan posisi.</p>
<p>Contoh Prefix :</p>
<p>+ A B</p>
<p>* - A B C</p>


<h2>POSTFiX</h2>
<p>Posfix adalah cara penulisan yang meletakan operator disebelah kanan 2 operand dan posisi operand yang berada di dalam kurung sangat menentukan.</p>
<p>Contoh Postfix :</p>
<p>A B +</p>
<p>A B - C *</p>
